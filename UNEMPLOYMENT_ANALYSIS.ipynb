```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset from a CSV file
file_path = 'Unemployment in India.csv'
df = pd.read_csv(file_path)
```

```python
# Display the loaded dataset
print("Loaded Unemployment Dataset:")
print(df)
```

```python
# Print column names
print(df.columns)
```

```python
# Remove leading and trailing whitespaces from column names
df.columns = df.columns.str.strip()
```

```python
# Handling missing values (replace with the average)
average_unemployment_rate = df['Estimated Unemployment Rate (%)'].mean()
df['Estimated Unemployment Rate (%)'].fillna(average_unemployment_rate, inplace=True)
```

```python
# Display the dataset after handling missing values
print("\nDataset after handling missing values:")
print(df)
```

```python
# Plotting the original and preprocessed data
plt.figure(figsize=(12, 6))
```

```python
# Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
# Check if conversion was successful
print(df['Date'])
```

```python
 #Convert 'Date' to string
df['Date'] = df['Date'].dt.strftime('%Y-%m-%d')
```

```python
# Check for NaN values in 'Date' column
nan_values = df['Date'].isna().any()

if nan_values:
    # Handle NaN values (replace with a placeholder or drop rows)
    df['Date'].fillna('Unknown', inplace=True)  # Replace NaN with 'Unknown' or choose a suitable placeholder
```

```python
# Convert 'Date' to string
df['Date'] = df['Date'].astype(str)

# Original data plot
plt.subplot(1, 2, 1)
plt.plot(df['Date'], df['Estimated Unemployment Rate (%)'], marker='o', linestyle='-', color='b')
plt.title('Unemployment Rate Over Time (Original)')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.grid(True)
```

```python
# Bar chart for average unemployment rate
plt.subplot(1, 2, 2)
plt.bar(df['Date'], df['Estimated Unemployment Rate (%)'], color='skyblue')
plt.axhline(y=average_unemployment_rate, color='red', linestyle='--', label='Average Rate')
plt.title('Average Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.legend()
plt.grid(True)
```

```python
plt.tight_layout()
plt.show()
```
